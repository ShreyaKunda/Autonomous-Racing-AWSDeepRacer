```python
#PPOR2-128-0.0003
import json

# Load the JSON data
file_path = "C:/AMRITA/Sem7/DRL/Endsem/PPO-R2-training_job_eyJonNQmQ9SFygzlu4GttQ_logs/4bd886f6-0748-4ff4-98f9-46a045520bd3/metrics/training/training-20241201061235-eyJonNQmQ9SFygzlu4GttQ.json"
with open(file_path, 'r') as file:
    data = json.load(file)

# Initialize variables for metrics
total_reward = 0
total_rewards_per_episode = []
episode_count = len(data['metrics'])
total_completions = 0
off_track_count = 0
total_time = 0
total_regret = 0
optimal_reward = 100  # Assuming an optimal reward (example value, adjust as needed)

# Calculate metrics
for episode in data['metrics']:
    reward_score = episode.get('reward_score', 0)
    total_reward += reward_score
    total_rewards_per_episode.append(reward_score)

    # Completion percentage
    completion_percentage = episode.get('completion_percentage', 0)
    total_completions += completion_percentage

    # Off-track episodes
    if episode.get('episode_status', '').lower() == 'off track':
        off_track_count += 1

    # Episode time (in seconds)
    elapsed_time_seconds = episode.get('elapsed_time_in_milliseconds', 0) / 1000
    total_time += elapsed_time_seconds

    # Regret calculation (optimal reward - achieved reward)
    regret = optimal_reward - reward_score
    total_regret += regret

# Calculate averages
average_reward = total_reward / episode_count if episode_count > 0 else 0
average_completion_percentage = total_completions / episode_count if episode_count > 0 else 0
average_time_per_episode = total_time / episode_count if episode_count > 0 else 0
average_regret = total_regret / episode_count if episode_count > 0 else 0

# Calculate additional metrics
success_rate = ((episode_count - off_track_count) / episode_count) * 100
cumulative_reward = total_reward
reward_per_second = total_reward / total_time if total_time > 0 else 0
percentage_completed_episodes = (episode_count - off_track_count) / episode_count * 100

# Print the metrics
print(f"Total Episodes: {episode_count}")
print(f"Total Reward: {total_reward}")
print(f"Average Reward per Episode: {average_reward:.2f}")
print(f"Average Completion Percentage: {average_completion_percentage:.2f}%")
print(f"Average Time per Episode (seconds): {average_time_per_episode:.2f}")
print(f"Number of Off-track Episodes: {off_track_count}")
print(f"Success Rate: {success_rate:.2f}%")
print(f"Cumulative Reward: {cumulative_reward}")
print(f"Regret (Average): {average_regret:.2f}")
print(f"Reward per Second: {reward_per_second:.2f}")
print(f"Percentage of Completed Episodes: {percentage_completed_episodes:.2f}%")
```
